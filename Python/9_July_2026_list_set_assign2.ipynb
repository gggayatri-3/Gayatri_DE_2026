```python
# Set Practice Questions
# Basic Operations

# 1.Print the customers set.
print(customers)
# 2.Find the number of elements in the customers set.
print(len(customers))
# 3.Check whether "Rahul" exists in the customers set.
print("Rahul" in customers)
# 4.Check whether "Meena" exists in the customers set.
print("Meena" in customers)
# 5.Iterate through all elements in the bank_branches set.
for branch in bank_branches:
    print(branch)
```

```python
# add()

# 1.Add "Meena" to the customers set.
customers.add("Meena")
print(customers)
# 2.Add "Mysore" to the cities set.
cities.add("Mysore")
print(cities)
```

```python
# update()

# Update the customers set with new_customers.
customers.update(new_customers)
print(customers)
# Update the cities set with new_cities.
cities.update(new_cities)
print(cities)
```

```python
# remove()

# Remove "Rahul" from the customers set.
customers.remove("Rahul")
print(customers)
# Remove "Pune" from the cities set.
cities.remove("Pune")
print(cities)
```

```python
# discard()

# Discard "Anita" from the customers set.
customers.discard("Anita")
print(customers)
# Discard "Delhi" from the cities set.
cities.discard("Delhi")
print(cities)
# Discard "Kolkata" from the cities set.
cities.discard("Kolkata")
print(cities)
```

```python
# pop()

# Remove a random element from the transaction_ids set using pop().
transaction_ids.pop()
print(transaction_ids)
# Remove another random element from the account_types set.
account_types.pop()
print(account_types)
```

```python
# clear()

# Remove all elements from the balances set.
balances.clear()
print(balances)
```

```python
# copy()

# Create a copy of the customers set.
customers_copy = customers.copy()
# Modify the copied set and verify that the original set remains unchanged.
customers_copy.add("Sunita")
print("Original set:", customers)
print("Copied set:", customers_copy)
```

```python
# union()

# Find the union of customers and new_customers.
union_set = customers.union(new_customers)
print("Union set:", union_set)
```

```python
# intersection()

# Find the common customers between customers and new_customers.
intersection_set = customers.intersection(new_customers)
print("Intersection set:", intersection_set)

# Find customers who have both Savings and Loan accounts.
common_customers = savings_customers.intersection(loan_customers)
print("Customers with both Savings and Loan accounts:", common_customers)
```

```python
# difference()

# Find customers present only in customers but not in new_customers.
difference_set = customers.difference(new_customers)
print("Customers present only in customers:", difference_set)

# Find customers who have only Savings accounts.
savings_only_customers = savings_customers.difference(loan_customers)
print("Customers with only Savings accounts:", savings_only_customers)
```

```python
# symmetric_difference()

# Find customers that are present in either customers or new_customers but not both.
symmetric_difference_set = customers.symmetric_difference(new_customers)
print("Customers present in either customers or new_customers but not both:", symmetric_difference_set)
```

```python
# issubset()

# Check whether {"Rahul", "Kiran"} is a subset of loan_customers.
is_subset = {"Rahul", "Kiran"}.issubset(loan_customers)
print("Is subset:", is_subset)
```

```python
# issuperset()

# Check whether customers is a superset of {"Rahul", "Anita"}.
is_superset = customers.issuperset({"Rahul", "Anita"})
print("Is superset:", is_superset)
```

```python
# sdisjoint()

# Check whether loan_customers and new_cities are disjoint.
is_disjoint = loan_customers.isdisjoint(new_cities)
print("Are disjoint:", is_disjoint)
```
